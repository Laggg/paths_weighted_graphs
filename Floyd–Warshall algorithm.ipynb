# Floyd–Warshall algorithm

Алгоритм находит оптимальный путь от каждой вершины до каждой в графе. Алгоритм хорош тем, что веса ребер могут быть отрицательными.
Допустим, что в графе нет циклов с отрицательной суммой весов ребер (иначе можно уйти в минус по циклу на сколько угодно, так как мы будем сравнивать числа).
Slozhost' algoritma - "эн куб"

```python
import numpy as np
```

```python
def FW(G):

    d = np.copy(G)
    for i in range(G.shape[0]+1):
        for j in range(G.shape[0]):
            for k in range(G.shape[0]):
                if d[j][k] > d[j][i - 1] + d[i - 1][k]:
                    d[j][k] = d[j][i - 1] + d[i - 1][k]
    
    #считаем, что нет петель от вершины на саму себя, т.е. ближайшее расстояние от вершины до самой себя = 0
    np.fill_diagonal(d, 0)
                
    return d
```

```python
#простейший пример работы с неориентированным графом, изображенным на рисунке (вершины: 0,1,2,3. Ребра: 03 = 2, 12 = 1, 13 = 1, 23 = 6). 
#Вес 100 там, где ребра нет (в идеале бесконечность).

G = np.array([[100,100,100,2],
     [100,100,1,1],
     [100,1,100,6],
     [2,1,6,100]])
print(G)
```

```python
FW(G)
```
